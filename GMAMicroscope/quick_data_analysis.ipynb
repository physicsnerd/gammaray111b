```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("D:/250823_073302_pulse_height_analyzer_histogram.csv", skiprows=15)
df.head()
```

```python
plt.hist(df["raw_heights"], bins=1024)
plt.show()
```

```python
# Plot pandas histogram from dataframe with df.plot.hist (not df.hist)
ax = df['raw_heights'].plot.hist(bins=2048, density=True)

# Save default x-axis limits for final formatting because the pandas kde
# plot uses much wider limits which usually decreases readability
xlim = ax.get_xlim()

# Plot pandas KDE
df['raw_heights'].plot.density(color='k', alpha=0.5, ax=ax) # same as df['var'].plot.kde()

# Reset x-axis limits and edit legend and add title
ax.set_xlim(xlim)
#ax.legend(labels=['KDE'], frameon=False)
ax.set_title('Counts vs Energy of Cs-137', fontsize=14, pad=15)
ax.set_xlabel('Voltage (V)')
ax.set_ylabel('Counts')

plt.show()
```
